```python
from scipy import linalg
import numpy as np
from matplotlib import pylab as plt
import math
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
f(1)
```

```python
a_linear = np.array([[1, 1], [1, 15]])
b_linear = np.array([f(1), f(15)])
print(a_linear, b_linear)
```

```python
x_linear = linalg.solve(a_linear, b_linear)
```

```python
print(x_linear)
```

```python
%matplotlib inline
```

```python
x = np.arange(1, 16, 1)
y = f(x)
y_linear = -0.19 * x +3.44  
plt.plot(x, y, '-', x, y_linear, '-')
plt.show()
```

```python
a_quadratic = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 15*15]])
b_quadratic = np.array([f(1), f(8), f(15)])
print(a_quadratic, b_quadratic)
```

```python
x_quadratic = linalg.solve(a_quadratic, b_quadratic)
x_quadratic
```

```python
x = np.arange(1, 16, 1)
y = f(x)
y_linear = -0.008 * x**2 -0.07 * x +3.33  
plt.plot(x, y, '-', x, y_linear, '-')
plt.show()
```

```python
a_qubic = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
b_qubic = np.array([f(1), f(4), f(10), f(15)])
print(a_qubic, b_qubic)
```

```python
x_qubic = linalg.solve(a_qubic, b_qubic)
x_qubic
```
