```python
f = open('sentences.txt', 'r')
data_text = []
for line in f:
    data_text.append(line.strip())
f.close()
```

```python
data_text
```

```python
import re
data_text_split = []
for line in data_text:
    data_text_split.append(re.split('[^a-z]', line.lower()))
```

```python
data_text_split
```

```python
clean_data = []
for sentence in data_text_split:
    for word in sentence:
        if word != '':
            clean_data.append(word) 
```

```python
dict_data = {}
i = 0
for word in clean_data:
    if word in dict_data.keys():
        continue
    else:
        dict_data[word] = i
        i += 1
```

```python
dict_data
```

```python
import numpy as np
n = len(data_text)
d = len(dict_data)

matrix = np.zeros((n, d))
```

```python
matrix
```

```python
matrix.shape
```

```python
from collections import Counter
i = 0
j = 0
for i in range(len(data_text_split)):
    for key in dict_data.keys():
        if key in data_text_split[i]:
            for word in data_text_split[i]:
                if word  == key:
                    j += 1
            matrix[i, dict_data.get(key)] += j
            j = 0
```

```python
matrix
```

```python
from scipy.spatial import distance as ds
```

```python
matrix[1]
```

```python
cosine = [] 
for line in matrix:
    cosine.append(ds.cosine(matrix[0], line))
    
cosine
```
